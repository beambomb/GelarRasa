Cek Missing Values, Duplicates, Outliers

```python
# Fungsi untuk cek data quality
def data_quality_check(df, name):
    print("="*60)
    print(f"DATA QUALITY CHECK: {name}")
    print("="*60)

    # Missing values
    print("\nMISSING VALUES:")
    missing = df.isnull().sum()
    missing_pct = (df.isnull().sum() / len(df)) * 100
    missing_df = pd.DataFrame({
        'Column': missing.index,
        'Missing_Count': missing.values,
        'Missing_Percentage': missing_pct.values
    })
    missing_df = missing_df[missing_df['Missing_Count'] > 0].sort_values('Missing_Count', ascending=False)

    if len(missing_df) > 0:
        print(missing_df.to_string(index=False))
    else:
        print("Tidak ada missing values!")

    # Duplicates
    print("\nDUPLICATE ROWS:")
    duplicates = df.duplicated().sum()
    print(f"Total duplicate rows: {duplicates}")
    if duplicates > 0:
        print(f"Persentase duplikat: {(duplicates/len(df)*100):.2f}%")
    else:
        print("Tidak ada duplikat!")

    print("\n" + "="*60 + "\n")

# Cek semua dataset
data_quality_check(products, "PRODUCTS")
data_quality_check(marketing, "MARKETING")
data_quality_check(reviews, "REVIEWS")
data_quality_check(sales, "SALES")
```

Merge Dataset untuk Analisis Komprehensif

```python
# Merge semua data menjadi satu master dataset
# Start dengan sales sebagai base (transaksi aktual)
master_df = sales.copy()

# Join dengan products
master_df = master_df.merge(products, on='product_id', how='left')

# Aggregate marketing per product (karena 1 product bisa punya multiple campaigns)
marketing_agg = marketing.groupby('product_id').agg({
    'campaign_id': 'count',  # Jumlah campaign
    'spend_idr': 'sum',          # Total marketing spend
    'engagement_rate': 'mean' # Average engagement rate
}).reset_index()

marketing_agg.columns = ['product_id', 'total_campaigns', 'total_marketing_spend', 'avg_engagement_rate']

master_df = master_df.merge(marketing_agg, on='product_id', how='left')

# Aggregate reviews per product
reviews_agg = reviews.groupby('product_id').agg({
    'rating': 'mean',        # Average rating
    'review_id': 'count'     # Jumlah reviews
}).reset_index()
reviews_agg.columns = ['product_id', 'avg_rating', 'review_count']

master_df = master_df.merge(reviews_agg, on='product_id', how='left')

print("="*50)
print("MASTER DATASET CREATED")
print("="*50)
print(f"Shape: {master_df.shape}")
print(f"\nColumns: {list(master_df.columns)}")
print(f"\nPreview:")
print(master_df.head())

# Save untuk penggunaan selanjutnya
master_df.to_csv('master_data.csv', index=False)
print("\nMaster dataset disimpan sebagai 'master_data.csv'")
```

# **BAGIAN 2: Data Cleaning & Preprocessing**

Identifikasi & Handle Missing Values

```python
# Cek missing values di master dataset
print("="*50)
print("HANDLING MISSING VALUES")
print("="*50)

missing_summary = master_df.isnull().sum()
missing_summary = missing_summary[missing_summary > 0].sort_values(ascending=False)

if len(missing_summary) > 0:
    print("\nKolom dengan missing values:")
    print(missing_summary)

    # Strategy untuk handle missing values
    print("\nSTRATEGI HANDLING:")

    # 1. Marketing-related columns (fill with 0 - produk tanpa campaign)
    marketing_cols = ['total_campaigns', 'total_marketing_spend', 'avg_engagement_rate']
    for col in marketing_cols:
        if col in master_df.columns and master_df[col].isnull().sum() > 0:
            master_df[col].fillna(0, inplace=True)
            print(f"{col}: Filled with 0 (produk tanpa marketing campaign)")

    # 2. Review-related columns
    if 'avg_rating' in master_df.columns and master_df['avg_rating'].isnull().sum() > 0:
        # Fill dengan median rating
        median_rating = master_df['avg_rating'].median()
        master_df['avg_rating'].fillna(median_rating, inplace=True)
        print(f"avg_rating: Filled with median ({median_rating:.2f})")

    if 'review_count' in master_df.columns and master_df['review_count'].isnull().sum() > 0:
        master_df['review_count'].fillna(0, inplace=True)
        print(f"review_count: Filled with 0 (produk tanpa review)")

    # 3. Numerical columns lain - fill dengan median
    numerical_cols = master_df.select_dtypes(include=[np.number]).columns
    for col in numerical_cols:
        if master_df[col].isnull().sum() > 0:
            median_val = master_df[col].median()
            master_df[col].fillna(median_val, inplace=True)
            print(f"{col}: Filled with median ({median_val:.2f})")

    # 4. Categorical columns - fill dengan mode atau 'Unknown'
    categorical_cols = master_df.select_dtypes(include=['object']).columns
    for col in categorical_cols:
        if master_df[col].isnull().sum() > 0:
            mode_val = master_df[col].mode()[0] if len(master_df[col].mode()) > 0 else 'Unknown'
            master_df[col].fillna(mode_val, inplace=True)
            print(f"{col}: Filled with mode ({mode_val})")

    print(f"\nSemua missing values berhasil ditangani!")
    print(f"Total missing values sekarang: {master_df.isnull().sum().sum()}")
else:
    print("Tidak ada missing values!")
```

Deteksi & Handle Outliers

```python
# Deteksi outliers menggunakan IQR method
print("="*50)
print("OUTLIER DETECTION & HANDLING")
print("="*50)

def detect_outliers_iqr(df, column):
    """Deteksi outliers menggunakan IQR method"""
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

    return outliers, lower_bound, upper_bound

# Kolom yang akan dicek outliers
outlier_cols = ['revenue', 'units_sold', 'discount_pct', 'base_price']

outlier_summary = {}

for col in outlier_cols:
    if col in master_df.columns:
        outliers, lower, upper = detect_outliers_iqr(master_df, col)
        outlier_summary[col] = {
            'count': len(outliers),
            'percentage': (len(outliers) / len(master_df)) * 100,
            'lower_bound': lower,
            'upper_bound': upper
        }

        print(f"\n{col.upper()}:")
        print(f"   Outliers: {len(outliers)} ({(len(outliers)/len(master_df)*100):.2f}%)")
        print(f"   Lower bound: {lower:.2f}")
        print(f"   Upper bound: {upper:.2f}")

# Visualisasi outliers dengan boxplot
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.ravel()

for idx, col in enumerate(outlier_cols[:4]):
    if col in master_df.columns:
        axes[idx].boxplot(master_df[col].dropna(), vert=False)
        axes[idx].set_title(f'Boxplot: {col}', fontsize=12, fontweight='bold')
        axes[idx].set_xlabel(col)

plt.tight_layout()
plt.show()
```

```python
# Strategi handling outliers
print("\nSTRATEGI HANDLING OUTLIERS:")
print("Untuk kompetisi data science, kita akan:")
print("1. CATAT outliers untuk analisis (mungkin insight berharga)")
print("2. BUAT FLAG column untuk menandai outliers")
print("3. TIDAK langsung hapus - karena bisa jadi transaksi legitimate")
print()

# Buat flag columns untuk outliers
for col in outlier_cols:
    if col in master_df.columns:
        outliers, lower, upper = detect_outliers_iqr(master_df, col)
        master_df[f'{col}_is_outlier'] = ((master_df[col] < lower) | (master_df[col] > upper)).astype(int)

print("Outlier flags created!")
print("\nContoh data dengan outlier flag:")
print(master_df[['revenue', 'revenue_is_outlier', 'units_sold', 'units_sold_is_outlier']].head(10))

# Optional: Capping extreme outliers (winsorization)
# Hanya untuk kolom yang benar-benar extreme
print("\n🔧 WINSORIZATION (Capping extreme values):")

def winsorize_column(df, column, lower_percentile=0.01, upper_percentile=0.99):
    """Cap extreme values pada percentile tertentu"""
    lower_cap = df[column].quantile(lower_percentile)
    upper_cap = df[column].quantile(upper_percentile)

    df[f'{column}_capped'] = df[column].clip(lower=lower_cap, upper=upper_cap)

    return df

# Terapkan winsorization untuk revenue (contoh)
master_df = winsorize_column(master_df, 'revenue', 0.01, 0.99)
print("Revenue winsorized (1st-99th percentile)")
```

Handle Duplicates

```python
print("="*50)
print("DATA TYPE VALIDATION & CONVERSION")
print("="*50)

# Pastikan tipe data sudah benar
print("\n Current data types:")
print(master_df.dtypes)

# Konversi yang mungkin diperlukan
conversions_made = []

# 1. Date columns
date_cols = ['date', 'launch_date']
for col in date_cols:
    if col in master_df.columns and master_df[col].dtype == 'object':
        master_df[col] = pd.to_datetime(master_df[col], errors='coerce')
        conversions_made.append(f"{col} → datetime")

# 2. Categorical columns yang seharusnya category type
categorical_cols = ['region', 'channel_name', 'brand', 'type', 'product_name']
for col in categorical_cols:
    if col in master_df.columns:
        master_df[col] = master_df[col].astype('category')
        conversions_made.append(f"{col} → category")

# 3. Numerical columns
numerical_cols = ['base_price', 'revenue', 'units_sold', 'discount_pct']
for col in numerical_cols:
    if col in master_df.columns and master_df[col].dtype == 'object':
        master_df[col] = pd.to_numeric(master_df[col], errors='coerce')
        conversions_made.append(f"{col} → numeric")

if conversions_made:
    print("\n Data type conversions:")
    for conversion in conversions_made:
        print(f"   - {conversion}")
else:
    print("\n All data types are correct!")

print("\n Updated data types:")
print(master_df.dtypes)
```

Step 1: Load Cleaned Data & Setup

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# Machine Learning libraries
from sklearn.model_selection import train_test_split, cross_val_score, TimeSeriesSplit
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, classification_report
import xgboost as xgb
from sklearn.model_selection import RandomizedSearchCV
import scipy.stats

# Load cleaned data
df = pd.read_csv('master_data_cleaned.csv')

# Convert date columns
if 'date' in df.columns:
    df['date'] = pd.to_datetime(df['date'])
if 'launch_date' in df.columns:
    df['launch_date'] = pd.to_datetime(df['launch_date'])

print("="*60)
print("DATA LOADED FOR FEATURE ENGINEERING")
print("="*60)
print(f"Shape: {df.shape}")
print(f"Columns: {list(df.columns)}")
print("\nFirst few rows:")
print(df.head())
```

Step 2: Business-Driven Feature Engineering

```python
print("\n" + "="*60)
print("FEATURE ENGINEERING - BUSINESS DRIVEN FEATURES")
print("="*60)

# ==========================================
# 1. SALES VELOCITY
# ==========================================
# Argumentasi: Mengukur seberapa cepat produk terjual sejak launch
# Formula: units_sold / days_since_launch
# Higher velocity = produk populer/fast-moving

if 'days_since_launch' in df.columns:
    # Pastikan tidak ada nilai 0 untuk menghindari division by zero
    df['days_since_launch'] = df['days_since_launch'].replace(0, 1)
    df['sales_velocity'] = df['units_sold'] / df['days_since_launch']
    print("Feature created: sales_velocity")
    print(f"   - Formula: units_sold / days_since_launch")
    print(f"   - Mean: {df['sales_velocity'].mean():.4f}")
    print(f"   - Business insight: Produk dengan velocity tinggi = fast-moving items")
else:
    print(" days_since_launch not found, creating from date difference...")
    if 'date' in df.columns and 'launch_date' in df.columns:
        df['days_since_launch'] = (df['date'] - df['launch_date']).dt.days
        df['days_since_launch'] = df['days_since_launch'].replace(0, 1)
        df['sales_velocity'] = df['units_sold'] / df['days_since_launch']
        print("Feature created: sales_velocity")

# ==========================================
# 2. MARKETING ROI
# ==========================================
# Argumentasi: Return on Investment dari marketing spend
# Formula: (revenue - marketing_spend) / marketing_spend
# Higher ROI = kampanye efektif

if 'total_marketing_spend' in df.columns:
    # Handle products without marketing (spend = 0)
    df['marketing_roi'] = np.where(
        df['total_marketing_spend'] > 0,
        (df['revenue'] - df['total_marketing_spend']) / df['total_marketing_spend'],
        0  # No marketing = ROI 0
    )
    print("\n Feature created: marketing_roi")
    print(f"   - Formula: (revenue - marketing_spend) / marketing_spend")
    print(f"   - Mean ROI: {df['marketing_roi'].mean():.2f}")
    print(f"   - Business insight: ROI > 0 = profitable campaign")

# ==========================================
# 3. REVENUE PER UNIT
# ==========================================
# Argumentasi: Average selling price per unit (after discount)
# Formula: revenue / units_sold
# Higher value = premium products or less discount

df['revenue_per_unit'] = df['revenue'] / df['units_sold'].replace(0, 1)
print("\n Feature created: revenue_per_unit")
print(f"   - Formula: revenue / units_sold")
print(f"   - Mean: Rp {df['revenue_per_unit'].mean():,.0f}")
print(f"   - Business insight: Reflects actual selling price after discount")

# ==========================================
# 4. EFFECTIVE DISCOUNT RATE
# ==========================================
# Argumentasi: Actual discount given in rupiah terms
# Formula: discount_pct * base_price / 100
# Shows real discount amount, not just percentage

if 'discount_pct' in df.columns and 'base_price' in df.columns:
    df['effective_discount_amount'] = (df['discount_pct'] * df['base_price']) / 100
    df['price_after_discount'] = df['base_price'] - df['effective_discount_amount']

    print("\n Feature created: effective_discount_amount")
    print(f"   - Formula: (discount_pct * base_price) / 100")
    print(f"   - Mean discount: Rp {df['effective_discount_amount'].mean():,.0f}")

# ==========================================
# 5. PROFIT MARGIN (Estimation)
# ==========================================
# Argumentasi: Estimated profit per transaction
# Assumption: Cost of goods = 60% of base_price (industry standard FMCG)
# Formula: revenue - (units_sold * base_price * 0.6) - marketing_spend

if 'base_price' in df.columns:
    df['estimated_cogs'] = df['units_sold'] * df['base_price'] * 0.6  # 60% COGS
    df['estimated_profit'] = df['revenue'] - df['estimated_cogs']

    if 'total_marketing_spend' in df.columns:
        df['estimated_profit'] = df['estimated_profit'] - df['total_marketing_spend']

    df['profit_margin_pct'] = (df['estimated_profit'] / df['revenue']) * 100

    print("\n Feature created: profit_margin_pct")
    print(f"   - Assumption: COGS = 60% of base_price")
    print(f"   - Mean profit margin: {df['profit_margin_pct'].mean():.2f}%")
    print(f"   - Business insight: Higher margin = more profitable products")

# ==========================================
# 6. CHANNEL PREFERENCE SCORE
# ==========================================
# Argumentasi: Revenue performance by channel relative to product average
# Formula: revenue / avg_revenue_for_product_in_channel

if 'channel_name' in df.columns and 'product_id' in df.columns:
    # Calculate average revenue per product per channel
    channel_product_avg = df.groupby(['product_id', 'channel_name'])['revenue'].transform('mean')
    product_overall_avg = df.groupby('product_id')['revenue'].transform('mean')

    df['channel_performance_ratio'] = channel_product_avg / product_overall_avg

    print("\n Feature created: channel_performance_ratio")
    print(f"   - Formula: channel_revenue / product_avg_revenue")
    print(f"   - Ratio > 1: Channel performs better than average for this product")

# ==========================================
# 7. REVIEW SENTIMENT IMPACT
# ==========================================
# Argumentasi: Impact of reviews on sales performance
# Categories: No reviews, Low rating, Medium, High rating

if 'avg_rating' in df.columns:
    df['review_category'] = pd.cut(
        df['avg_rating'],
        bins=[0, 2.5, 3.5, 4.5, 5.0],
        labels=['Poor', 'Average', 'Good', 'Excellent']
    )

    # Binary: Has reviews or not
    df['has_reviews'] = (df['review_count'] > 0).astype(int)

    print("\n Feature created: review_category & has_reviews")
    print(f"   - Categories: Poor, Average, Good, Excellent")
    print(f"   - Products with reviews: {df['has_reviews'].sum()} ({df['has_reviews'].mean()*100:.1f}%)")

# ==========================================
# 8. PRICE POSITIONING
# ==========================================
# Argumentasi: Product positioning relative to category pricing
# Formula: product_price / avg_category_price

if 'type' in df.columns and 'base_price' in df.columns:
    category_avg_price = df.groupby('type')['base_price'].transform('mean')
    df['price_position_ratio'] = df['base_price'] / category_avg_price

    df['price_segment'] = pd.cut(
        df['price_position_ratio'],
        bins=[0, 0.8, 1.2, np.inf],
        labels=['Budget', 'Mid-range', 'Premium']
    )

    print("\n Feature created: price_position_ratio & price_segment")
    print(f"   - Budget: <80% of category average")
    print(f"   - Mid-range: 80-120% of category average")
    print(f"   - Premium: >120% of category average")

# ==========================================
# 9. PRODUCT LIFECYCLE STAGE
# ==========================================
# Argumentasi: Product maturity affects sales patterns
# Categories: Launch (0-90 days), Growth (91-365), Mature (366-730), Decline (>730)

if 'days_since_launch' in df.columns:
    df['lifecycle_stage'] = pd.cut(
        df['days_since_launch'],
        bins=[0, 90, 365, 730, np.inf],
        labels=['Launch', 'Growth', 'Mature', 'Decline']
    )

    print("\n Feature created: lifecycle_stage")
    print(f"   - Launch: 0-90 days, Growth: 91-365, Mature: 366-730, Decline: >730")
    print("\nDistribution:")
    print(df['lifecycle_stage'].value_counts())

# ==========================================
# 10. SEASONALITY FEATURES
# ==========================================
# Argumentasi: FMCG sales highly influenced by seasons, holidays, month-end

if 'date' in df.columns:
    df['month'] = df['date'].dt.month
    df['quarter'] = df['date'].dt.quarter
    df['day_of_week'] = df['date'].dt.dayofweek
    df['is_weekend'] = (df['day_of_week'] >= 5).astype(int)
    df['is_month_start'] = (df['date'].dt.day <= 7).astype(int)
    df['is_month_end'] = (df['date'].dt.day >= 24).astype(int)

    # High season (Q4: Oct-Dec) for FMCG
    df['is_high_season'] = (df['quarter'] == 4).astype(int)

    print("\n Features created: month, quarter, day_of_week, is_weekend")
    print(f"   - Weekend transactions: {df['is_weekend'].sum()} ({df['is_weekend'].mean()*100:.1f}%)")
    print(f"   - High season (Q4): {df['is_high_season'].sum()} ({df['is_high_season'].mean()*100:.1f}%)")

# ==========================================
# 11. ENGAGEMENT EFFICIENCY
# ==========================================
# Argumentasi: How well marketing engagement converts to sales
# Formula: units_sold / (marketing_spend * engagement_rate)

if 'total_marketing_spend' in df.columns and 'avg_engagement_rate' in df.columns:
    df['engagement_efficiency'] = np.where(
        (df['total_marketing_spend'] > 0) & (df['avg_engagement_rate'] > 0),
        df['units_sold'] / (df['total_marketing_spend'] * df['avg_engagement_rate']),
        0
    )

    print("\n Feature created: engagement_efficiency")
    print(f"   - Measures: conversion from engagement to actual sales")

# ==========================================
# 12. REGIONAL PERFORMANCE INDEX
# ==========================================
# Argumentasi: Product performance relative to region average

if 'region' in df.columns:
    region_avg_revenue = df.groupby('region')['revenue'].transform('mean')
    df['regional_performance_index'] = df['revenue'] / region_avg_revenue

    print("\n Feature created: regional_performance_index")
    print(f"   - Index > 1: Above-average performance in region")

print("\n" + "="*60)
print(f"TOTAL NEW FEATURES CREATED: {len(df.columns) - len(pd.read_csv('master_data_cleaned.csv').columns)}")
print("="*60)
```

Step 3: Interaction Features (Advanced)

```python
print("\n" + "="*60)
print("ADVANCED FEATURE ENGINEERING - INTERACTIONS")
print("="*60)

# ==========================================
# 13. DISCOUNT × RATING INTERACTION
# ==========================================
# Argumentasi: Produk high-rated mungkin tidak perlu banyak diskon
# Negative correlation expected

if 'discount_pct' in df.columns and 'avg_rating' in df.columns:
    df['discount_rating_interaction'] = df['discount_pct'] * df['avg_rating']
    print(" Feature: discount_rating_interaction")
    print("   - Hypothesis: High-rated products need less discount")

# ==========================================
# 14. PRICE × CHANNEL INTERACTION
# ==========================================
# Argumentasi: Different channels perform differently for different price points

if 'base_price' in df.columns and 'channel_name' in df.columns:
    # Normalize price first
    price_normalized = (df['base_price'] - df['base_price'].mean()) / df['base_price'].std()

    # Create dummy for top 3 channels
    top_channels = df['channel_name'].value_counts().head(3).index
    for channel in top_channels:
        df[f'price_x_{channel}'] = price_normalized * (df['channel_name'] == channel).astype(int)

    print(f" Features: price_x_[top_3_channels]")

# ==========================================
# 15. MARKETING × LIFECYCLE INTERACTION
# ==========================================
# Argumentasi: Marketing effectiveness varies by product lifecycle

if 'total_marketing_spend' in df.columns and 'lifecycle_stage' in df.columns:
    for stage in df['lifecycle_stage'].cat.categories:
        df[f'marketing_in_{stage}'] = df['total_marketing_spend'] * (df['lifecycle_stage'] == stage).astype(int)

    print(f" Features: marketing_in_[lifecycle_stages]")

print("\n Interaction features created!")
```

Step 4: Encoding Categorical Variables

```python
print("\n" + "="*60)
print("ENCODING CATEGORICAL VARIABLES")
print("="*60)

# Identify categorical columns
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

# Remove columns that shouldn't be encoded (IDs, dates)
exclude_cols = ['transaction_id', 'product_id', 'campaign_id', 'date', 'launch_date']
categorical_cols = [col for col in categorical_cols if col not in exclude_cols]

print(f"\nCategorical columns to encode: {categorical_cols}")

# Label Encoding for ordinal features
ordinal_features = {
    'review_category': {'Poor': 0, 'Average': 1, 'Good': 2, 'Excellent': 3},
    'price_segment': {'Budget': 0, 'Mid-range': 1, 'Premium': 2},
    'lifecycle_stage': {'Launch': 0, 'Growth': 1, 'Mature': 2, 'Decline': 3}
}

for col, mapping in ordinal_features.items():
    if col in df.columns:
        df[f'{col}_encoded'] = df[col].map(mapping)
        print(f" Label encoded: {col}")

# One-Hot Encoding for nominal features
nominal_features = ['region', 'channel_name', 'brand', 'type']
nominal_features = [col for col in nominal_features if col in df.columns]

if nominal_features:
    df_encoded = pd.get_dummies(df, columns=nominal_features, prefix=nominal_features, drop_first=True)
    print(f"\n One-hot encoded: {nominal_features}")
    print(f"New shape after encoding: {df_encoded.shape}")
else:
    df_encoded = df.copy()

print("\n Encoding completed!")
```

Step 5: Feature Scaling & Selection

```python
print("\n" + "="*60)
print("FEATURE SCALING & SELECTION")
print("="*60)

# Separate features and target
# Remove non-feature columns
drop_cols = ['transaction_id', 'date', 'launch_date', 'product_id', 'campaign_id',
             'review_category', 'price_segment', 'lifecycle_stage']  # Keep encoded versions
drop_cols = [col for col in drop_cols if col in df_encoded.columns]

# Define target variable (we'll create multiple targets for different tasks)
# Target 1: Revenue (Regression)
# Target 2: High/Low performer (Classification)

# Create classification target: High performer = revenue > 75th percentile
revenue_threshold = df_encoded['revenue'].quantile(0.75)
df_encoded['high_performer'] = (df_encoded['revenue'] > revenue_threshold).astype(int)

print(f"\nClassification target created: high_performer")
print(f"Threshold: Rp {revenue_threshold:,.0f}")
print(f"Distribution: {df_encoded['high_performer'].value_counts().to_dict()}")

# Select numerical features for scaling
numerical_features = df_encoded.select_dtypes(include=[np.number]).columns.tolist()

# Remove targets and IDs from feature list
exclude_from_features = ['revenue', 'high_performer', 'units_sold']  # units_sold might be target too
numerical_features = [col for col in numerical_features if col not in exclude_from_features]

print(f"\nTotal features for modeling: {len(numerical_features)}")

# Save feature-engineered dataset
df_encoded.to_csv('data_with_features.csv', index=False)
print("\n Feature-engineered dataset saved as 'data_with_features.csv'")
```

Step 6: Define Modeling Tasks

```python
print("\n" + "="*60)
print("DEFINING MODELING TASKS")
print("="*60)

# Task 1: INNOVATION RADAR - Identify high-potential products
# Type: Binary Classification
# Target: high_performer (1 = revenue > 75th percentile)
# Goal: Predict which products will be top performers

print("\n TASK 1: INNOVATION RADAR (Classification)")
print("   - Objective: Identify high-potential products")
print("   - Target: high_performer (binary)")
print("   - Business value: Focus R&D and marketing on high-potential products")

# Task 2: TREND FORECASTING - Predict future revenue
# Type: Regression
# Target: revenue
# Goal: Forecast sales revenue for inventory and production planning

print("\n TASK 2: TREND FORECASTING (Regression)")
print("   - Objective: Predict product revenue")
print("   - Target: revenue (continuous)")
print("   - Business value: Optimize inventory, production, and supply chain")

# Task 3: PRODUCT CANNIBALIZATION (will do in Day 3)
print("\n TASK 3: PRODUCT CANNIBALIZATION (Day 3)")
print("   - Objective: Identify products that compete with each other")
print("   - Method: Correlation analysis, market basket analysis")
print("   - Business value: Optimize product portfolio")

print("\nWe'll focus on Task 1 & 2 today!")
```

Step 7: Data Preparation for Modeling

```python
print("\n" + "="*60)
print("DATA PREPARATION FOR MODELING")
print("="*60)

# Prepare feature matrix X and targets y
X = df_encoded[numerical_features].copy()

# Handle any remaining missing values
X = X.fillna(X.median())

# Handle infinite values
X = X.replace([np.inf, -np.inf], np.nan)
X = X.fillna(X.median())

# Targets
y_regression = df_encoded['revenue'].copy()
y_classification = df_encoded['high_performer'].copy()

print(f" Feature matrix X shape: {X.shape}")
print(f" Regression target shape: {y_regression.shape}")
print(f" Classification target shape: {y_classification.shape}")

# Train-test split
# For time series data, use temporal split
if 'date' in df_encoded.columns:
    print("\n Using temporal train-test split (last 20% as test)")

    # Sort by date
    df_encoded_sorted = df_encoded.sort_values('date').reset_index(drop=True)
    X_sorted = X.loc[df_encoded_sorted.index]
    y_reg_sorted = y_regression.loc[df_encoded_sorted.index]
    y_clf_sorted = y_classification.loc[df_encoded_sorted.index]

    split_idx = int(len(X_sorted) * 0.8)

    X_train = X_sorted[:split_idx]
    X_test = X_sorted[split_idx:]
    y_reg_train = y_reg_sorted[:split_idx]
    y_reg_test = y_reg_sorted[split_idx:]
    y_clf_train = y_clf_sorted[:split_idx]
    y_clf_test = y_clf_sorted[split_idx:]

else:
    print("\n Using random train-test split (80-20)")
    X_train, X_test, y_reg_train, y_reg_test = train_test_split(
        X, y_regression, test_size=0.2, random_state=42
    )
    X_train, X_test, y_clf_train, y_clf_test = train_test_split(
        X, y_classification, test_size=0.2, random_state=42
    )

print(f"\nTrain set size: {X_train.shape[0]:,}")
print(f"Test set size: {X_test.shape[0]:,}")

# Feature Scaling (StandardScaler)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Convert back to DataFrame for easier handling
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns, index=X_train.index)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns, index=X_test.index)

print("\n Feature scaling completed using StandardScaler")
```

Step 8: Baseline Model - Task 1 (Innovation Radar - Classification)

```python
print("\n" + "="*60)
print("TASK 1: INNOVATION RADAR - BASELINE MODEL")
print("="*60)

# Model 1: Random Forest Classifier (Baseline)
print("\n Training Random Forest Classifier...")

rf_clf = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    min_samples_split=20,
    min_samples_leaf=10,
    random_state=42,
    n_jobs=-1
)

rf_clf.fit(X_train_scaled, y_clf_train)

# Predictions
y_clf_pred_train = rf_clf.predict(X_train_scaled)
y_clf_pred_test = rf_clf.predict(X_test_scaled)
y_clf_pred_proba = rf_clf.predict_proba(X_test_scaled)[:, 1]

# Evaluation
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix

print("\n RANDOM FOREST CLASSIFIER - RESULTS:")
print(f"Train Accuracy: {accuracy_score(y_clf_train, y_clf_pred_train):.4f}")
print(f"Test Accuracy: {accuracy_score(y_clf_test, y_clf_pred_test):.4f}")
print(f"\nTest Metrics:")
print(f"  Precision: {precision_score(y_clf_test, y_clf_pred_test):.4f}")
print(f"  Recall: {recall_score(y_clf_test, y_clf_pred_test):.4f}")
print(f"  F1-Score: {f1_score(y_clf_test, y_clf_pred_test):.4f}")
print(f"  ROC-AUC: {roc_auc_score(y_clf_test, y_clf_pred_proba):.4f}")

print(f"\nConfusion Matrix:")
cm = confusion_matrix(y_clf_test, y_clf_pred_test)
print(cm)

# Feature Importance
feature_importance = pd.DataFrame({
    'feature': X_train.columns,
    'importance': rf_clf.feature_importances_
}).sort_values('importance', ascending=False)

print("\n TOP 15 MOST IMPORTANT FEATURES:")
print(feature_importance.head(15).to_string(index=False))

# Visualize feature importance
plt.figure(figsize=(10, 8))
plt.barh(feature_importance.head(15)['feature'], feature_importance.head(15)['importance'])
plt.xlabel('Importance Score')
plt.title('Top 15 Feature Importance - Random Forest Classifier', fontweight='bold')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

print("\n BUSINESS INTERPRETATION:")
print("The model can predict high-performing products with reasonable accuracy.")
print("Use these predictions to:")
print("  - Prioritize R&D resources on high-potential products")
print("  - Allocate marketing budget to products with growth potential")
print("  - Optimize inventory for predicted high-performers")
```

Step 9: Advanced Model - Task 1 (XGBoost with Hyperparameter Tuning)

```python
print("\n" + "="*60)
print("TASK 1: INNOVATION RADAR - ADVANCED MODEL (XGBoost)")
print("="*60)

# Model 2: XGBoost Classifier dengan Hyperparameter Tuning
print("\n Training XGBoost Classifier with Hyperparameter Tuning...")

# Define parameter distributions for RandomizedSearchCV
param_distributions = {
    'n_estimators': [100, 200, 300, 500],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'max_depth': [3, 5, 7, 10],
    'min_child_weight': [1, 3, 5],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'gamma': [0, 0.1, 0.2, 0.3],
    'reg_alpha': [0, 0.01, 0.1, 1],
    'reg_lambda': [0, 0.01, 0.1, 1]
}

# Base XGBoost model
xgb_clf_base = xgb.XGBClassifier(
    random_state=42,
    n_jobs=-1,
    eval_metric='logloss'
)

# Randomized Search
print("Running RandomizedSearchCV (this may take a few minutes)...")

random_search = RandomizedSearchCV(
    estimator=xgb_clf_base,
    param_distributions=param_distributions,
    n_iter=50,  # Number of parameter settings sampled
    scoring='roc_auc',
    cv=5,
    verbose=1,
    random_state=42,
    n_jobs=-1
)

random_search.fit(X_train_scaled, y_clf_train)

print(f"\n Best Parameters Found:")
best_params = random_search.best_params_
for param, value in best_params.items():
    print(f"   {param}: {value}")

print(f"\n Best CV ROC-AUC Score: {random_search.best_score_:.4f}")

# Train final model with best parameters
xgb_clf = random_search.best_estimator_

# Predictions
y_clf_pred_xgb_train = xgb_clf.predict(X_train_scaled)
y_clf_pred_xgb_test = xgb_clf.predict(X_test_scaled)
y_clf_pred_xgb_proba = xgb_clf.predict_proba(X_test_scaled)[:, 1]

# Evaluation
print("\n XGBOOST CLASSIFIER - RESULTS:")
print(f"Train Accuracy: {accuracy_score(y_clf_train, y_clf_pred_xgb_train):.4f}")
print(f"Test Accuracy: {accuracy_score(y_clf_test, y_clf_pred_xgb_test):.4f}")
print(f"\nTest Metrics:")
print(f"  Precision: {precision_score(y_clf_test, y_clf_pred_xgb_test):.4f}")
print(f"  Recall: {recall_score(y_clf_test, y_clf_pred_xgb_test):.4f}")
print(f"  F1-Score: {f1_score(y_clf_test, y_clf_pred_xgb_test):.4f}")
print(f"  ROC-AUC: {roc_auc_score(y_clf_test, y_clf_pred_xgb_proba):.4f}")

print("\n COMPARISON: Random Forest vs XGBoost")
print(f"Random Forest Test Accuracy: {accuracy_score(y_clf_test, y_clf_pred_test):.4f}")
print(f"XGBoost Test Accuracy: {accuracy_score(y_clf_test, y_clf_pred_xgb_test):.4f}")
print(f"\nRandom Forest ROC-AUC: {roc_auc_score(y_clf_test, y_clf_pred_proba):.4f}")
print(f"XGBoost ROC-AUC: {roc_auc_score(y_clf_test, y_clf_pred_xgb_proba):.4f}")

# Feature Importance dari XGBoost
feature_importance_xgb = pd.DataFrame({
    'feature': X_train.columns,
    'importance': xgb_clf.feature_importances_
}).sort_values('importance', ascending=False)

print("\n TOP 15 FEATURES (XGBoost):")
print(feature_importance_xgb.head(15).to_string(index=False))
```

Step 1: Load Models & Data

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Ensemble & Advanced ML
from sklearn.ensemble import StackingClassifier, StackingRegressor, VotingClassifier, VotingRegressor
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LogisticRegression, LinearRegression, Ridge
from sklearn.svm import SVC, SVR
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
import xgboost as xgb
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, roc_auc_score, mean_squared_error, r2_score
import joblib

# Load data
df = pd.read_csv('data_with_features.csv')

# Load previously trained models
try:
    xgb_clf = joblib.load('xgb_innovation_radar_model.pkl')
    xgb_reg = joblib.load('xgb_revenue_forecasting_model.pkl')
    scaler = joblib.load('feature_scaler.pkl')
    print(" Previous models loaded successfully!")
except:
    print(" Previous models not found. Please run Day 2 code first.")

# Prepare data (same as Day 2)
drop_cols = ['transaction_id', 'date', 'launch_date', 'product_id', 'campaign_id']
drop_cols = [col for col in drop_cols if col in df.columns]

numerical_features = df.select_dtypes(include=[np.number]).columns.tolist()
exclude_from_features = ['revenue', 'high_performer', 'units_sold']
numerical_features = [col for col in numerical_features if col not in exclude_from_features]

X = df[numerical_features].fillna(0).replace([np.inf, -np.inf], 0)
y_classification = df['high_performer']
y_regression = df['revenue']

# Train-test split (temporal)
if 'date' in df.columns:
    df_sorted = df.sort_values('date').reset_index(drop=True)
    X_sorted = X.loc[df_sorted.index]
    y_clf_sorted = y_classification.loc[df_sorted.index]
    y_reg_sorted = y_regression.loc[df_sorted.index]

    split_idx = int(len(X_sorted) * 0.8)
    X_train, X_test = X_sorted[:split_idx], X_sorted[split_idx:]
    y_clf_train, y_clf_test = y_clf_sorted[:split_idx], y_clf_sorted[split_idx:]
    y_reg_train, y_reg_test = y_reg_sorted[:split_idx], y_reg_sorted[split_idx:]
else:
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_clf_train, y_clf_test = train_test_split(X, y_classification, test_size=0.2, random_state=42)
    X_train, X_test, y_reg_train, y_reg_test = train_test_split(X, y_regression, test_size=0.2, random_state=42)

# Scale features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"\nData prepared: {X_train.shape[0]} train samples, {X_test.shape[0]} test samples")
```

Step 2: Stacking Ensemble for Classification (Innovation Radar)

```python
print("\n" + "="*60)
print("ADVANCED TECHNIQUE 1: STACKING ENSEMBLE - CLASSIFICATION")
print("="*60)

# Define base models (Level 0)
print("\n🔧 Building Stacking Ensemble...")

base_models_clf = [
    ('rf', RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42, n_jobs=-1)),
    ('gb', GradientBoostingClassifier(n_estimators=100, max_depth=5, random_state=42)),
    ('xgb', xgb.XGBClassifier(n_estimators=100, max_depth=5, random_state=42, n_jobs=-1, eval_metric='logloss')),
    ('knn', KNeighborsClassifier(n_neighbors=5))
]

# Define meta-model (Level 1)
meta_model_clf = LogisticRegression(max_iter=1000, random_state=42)

# Create Stacking Classifier
stacking_clf = StackingClassifier(
    estimators=base_models_clf,
    final_estimator=meta_model_clf,
    cv=5,  # 5-fold cross-validation for base models
    stack_method='predict_proba',  # Use probabilities for meta-model
    n_jobs=-1
)

print("📊 Base Models:")
for name, model in base_models_clf:
    print(f"   - {name.upper()}: {model.__class__.__name__}")
print(f"\n🎯 Meta-Model: {meta_model_clf.__class__.__name__}")

# Train stacking ensemble
print("\n⏳ Training stacking ensemble (this may take a few minutes)...")
stacking_clf.fit(X_train_scaled, y_clf_train)

# Predictions
y_clf_pred_stack = stacking_clf.predict(X_test_scaled)
y_clf_pred_stack_proba = stacking_clf.predict_proba(X_test_scaled)[:, 1]

# Evaluation
from sklearn.metrics import classification_report, confusion_matrix

print("\n" + "="*60)
print("STACKING ENSEMBLE - RESULTS")
print("="*60)

print(f"\n📊 Test Accuracy: {accuracy_score(y_clf_test, y_clf_pred_stack):.4f}")
print(f"📊 Test ROC-AUC: {roc_auc_score(y_clf_test, y_clf_pred_stack_proba):.4f}")

print("\nClassification Report:")
print(classification_report(y_clf_test, y_clf_pred_stack))

print("\nConfusion Matrix:")
cm_stack = confusion_matrix(y_clf_test, y_clf_pred_stack)
print(cm_stack)

# Compare with previous best model (XGBoost)
print("\n" + "="*60)
print("MODEL COMPARISON")
print("="*60)

# Evaluate individual base models
base_scores = {}
for name, model in base_models_clf:
    model.fit(X_train_scaled, y_clf_train)
    y_pred_proba = model.predict_proba(X_test_scaled)[:, 1]
    roc_auc = roc_auc_score(y_clf_test, y_pred_proba)
    base_scores[name] = roc_auc
    print(f"{name.upper()} ROC-AUC: {roc_auc:.4f}")

print(f"\n🏆 STACKING ENSEMBLE ROC-AUC: {roc_auc_score(y_clf_test, y_clf_pred_stack_proba):.4f}")

# Visualize comparison
models = list(base_scores.keys()) + ['Stacking']
scores = list(base_scores.values()) + [roc_auc_score(y_clf_test, y_clf_pred_stack_proba)]

plt.figure(figsize=(10, 6))
bars = plt.bar(models, scores, color=['steelblue', 'coral', 'green', 'purple', 'gold'])
plt.ylabel('ROC-AUC Score')
plt.title('Model Comparison: Stacking vs Base Models', fontweight='bold', fontsize=14)
plt.ylim([min(scores) - 0.02, 1.0])
plt.axhline(y=max(scores), color='red', linestyle='--', linewidth=1, label='Best Score')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print("\n💡 BUSINESS INTERPRETATION:")
print(f"Stacking ensemble combines the strengths of {len(base_models_clf)} different models.")
print(f"The ensemble achieves {roc_auc_score(y_clf_test, y_clf_pred_stack_proba):.4f} ROC-AUC, which is")
improvement = (roc_auc_score(y_clf_test, y_clf_pred_stack_proba) - max(base_scores.values())) * 100
print(f"{improvement:.2f}% {'better' if improvement > 0 else 'same as'} the best individual model.")
print("This robust prediction helps identify high-potential products with higher confidence.")

# Save stacking model
joblib.dump(stacking_clf, 'stacking_innovation_radar_model.pkl')
print("\n✅ Stacking model saved as 'stacking_innovation_radar_model.pkl'")
```

Step 3: Stacking Ensemble for Regression (Trend Forecasting)

```python
print("\n" + "="*60)
print("STACKING ENSEMBLE - REGRESSION (REVENUE FORECASTING)")
print("="*60)

# Define base models for regression
base_models_reg = [
    ('rf', RandomForestRegressor(n_estimators=100, max_depth=15, random_state=42, n_jobs=-1)),
    ('gb', GradientBoostingRegressor(n_estimators=100, max_depth=5, random_state=42)),
    ('xgb', xgb.XGBRegressor(n_estimators=100, max_depth=5, random_state=42, n_jobs=-1)),
    ('knn', KNeighborsRegressor(n_neighbors=5))
]

# Meta-model
meta_model_reg = Ridge(alpha=1.0, random_state=42)

# Create Stacking Regressor
stacking_reg = StackingRegressor(
    estimators=base_models_reg,
    final_estimator=meta_model_reg,
    cv=5,
    n_jobs=-1
)

print(" Base Models:")
for name, model in base_models_reg:
    print(f"   - {name.upper()}: {model.__class__.__name__}")
print(f"\n Meta-Model: {meta_model_reg.__class__.__name__}")

# Train
print("\n Training stacking regressor...")
stacking_reg.fit(X_train_scaled, y_reg_train)

# Predictions
y_reg_pred_stack = stacking_reg.predict(X_test_scaled)

# Evaluation
print("\n" + "="*60)
print("STACKING REGRESSOR - RESULTS")
print("="*60)

rmse_stack = np.sqrt(mean_squared_error(y_reg_test, y_reg_pred_stack))
r2_stack = r2_score(y_reg_test, y_reg_pred_stack)
mape_stack = np.mean(np.abs((y_reg_test - y_reg_pred_stack) / y_reg_test)) * 100

print(f"\n Test R² Score: {r2_stack:.4f}")
print(f" Test RMSE: Rp {rmse_stack:,.0f}")
print(f" Test MAPE: {mape_stack:.2f}%")

# Compare with base models
print("\n" + "="*60)
print("REGRESSION MODEL COMPARISON")
print("="*60)

base_rmse = {}
for name, model in base_models_reg:
    model.fit(X_train_scaled, y_reg_train)
    y_pred = model.predict(X_test_scaled)
    rmse = np.sqrt(mean_squared_error(y_reg_test, y_pred))
    base_rmse[name] = rmse
    print(f"{name.upper()} RMSE: Rp {rmse:,.0f}")

print(f"\n STACKING ENSEMBLE RMSE: Rp {rmse_stack:,.0f}")

# Visualize predictions
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Actual vs Predicted
axes[0].scatter(y_reg_test, y_reg_pred_stack, alpha=0.5, s=20, color='darkgreen')
axes[0].plot([y_reg_test.min(), y_reg_test.max()], [y_reg_test.min(), y_reg_test.max()], 'r--', lw=2)
axes[0].set_xlabel('Actual Revenue (IDR)', fontsize=12)
axes[0].set_ylabel('Predicted Revenue (IDR)', fontsize=12)
axes[0].set_title('Stacking Ensemble: Actual vs Predicted', fontweight='bold', fontsize=14)
axes[0].grid(alpha=0.3)

# Residual plot
residuals = y_reg_test - y_reg_pred_stack
axes[1].scatter(y_reg_pred_stack, residuals, alpha=0.5, s=20, color='darkgreen')
axes[1].axhline(y=0, color='r', linestyle='--', lw=2)
axes[1].set_xlabel('Predicted Revenue (IDR)', fontsize=12)
axes[1].set_ylabel('Residuals (IDR)', fontsize=12)
axes[1].set_title('Residual Plot - Stacking Ensemble', fontweight='bold', fontsize=14)
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

print("\n💡 BUSINESS INTERPRETATION:")
print(f"Stacking ensemble achieves RMSE of Rp {rmse_stack:,.0f} ({mape_stack:.1f}% MAPE)")
print("This improved accuracy enables:")
print("  - More precise inventory planning (reduce overstock/stockout)")
print("  - Better budget allocation for production")
print("  - Accurate revenue projections for financial planning")

# Save model
joblib.dump(stacking_reg, 'stacking_revenue_forecasting_model.pkl')
print("\n Stacking regressor saved!")
```

###BAGIAN 2: PRODUCT CANNIBALIZATION ANALYSIS

Step 4: Cannibalization Rate Calculation

```python
print("\n" + "="*60)
print("ADVANCED TECHNIQUE 2: PRODUCT CANNIBALIZATION ANALYSIS")
print("="*60)

# Load original data with product information
df_full = pd.read_csv('data_with_features.csv')

# Ensure we have necessary columns
required_cols = ['product_id', 'product_name', 'brand', 'revenue', 'date']
if not all(col in df_full.columns for col in required_cols):
    print(" Missing required columns for cannibalization analysis")
else:
    print(" Data loaded for cannibalization analysis")

    # Convert date
    df_full['date'] = pd.to_datetime(df_full['date'])

    # Step 1: Calculate sales before and after product launches
    print("\n STEP 1: Analyzing sales trends by product...")

    # Group by product and time period
    product_sales = df_full.groupby(['product_id', 'product_name', 'brand']).agg({
        'revenue': 'sum',
        'units_sold': 'sum',
        'date': ['min', 'max']
    }).reset_index()

    product_sales.columns = ['product_id', 'product_name', 'brand', 'total_revenue',
                              'total_units', 'first_sale', 'last_sale']

    print(f"\nTotal products analyzed: {len(product_sales)}")
    print("\nTop 10 products by revenue:")
    print(product_sales.nlargest(10, 'total_revenue')[['product_name', 'brand', 'total_revenue']].to_string(index=False))

    # Step 2: Identify product pairs within same brand/category
    print("\n STEP 2: Identifying potential cannibalization pairs...")

    # Focus on products within same brand
    cannibalization_pairs = []

    for brand in df_full['brand'].unique():
        brand_products = product_sales[product_sales['brand'] == brand]

        if len(brand_products) > 1:
            # For each product pair in the brand
            for i, prod1 in brand_products.iterrows():
                for j, prod2 in brand_products.iterrows():
                    if i < j:  # Avoid duplicates
                        # Check if launch dates overlap (newer product launched while older exists)
                        if prod2['first_sale'] > prod1['first_sale']:
                            cannibalization_pairs.append({
                                'brand': brand,
                                'existing_product': prod1['product_name'],
                                'existing_product_id': prod1['product_id'],
                                'new_product': prod2['product_name'],
                                'new_product_id': prod2['product_id'],
                                'existing_launch': prod1['first_sale'],
                                'new_launch': prod2['first_sale']
                            })

    cannib_df = pd.DataFrame(cannibalization_pairs)
    print(f"\nIdentified {len(cannib_df)} potential cannibalization pairs")

    # Step 3: Calculate cannibalization rate
    print("\n STEP 3: Calculating cannibalization rates...")

    # For each pair, calculate sales before and after new product launch
    cannib_results = []

    for idx, pair in cannib_df.iterrows():
        # Get sales of existing product before new product launch
        existing_before = df_full[
            (df_full['product_id'] == pair['existing_product_id']) &
            (df_full['date'] < pair['new_launch'])
        ]['revenue'].sum()

        # Get sales of existing product after new product launch
        existing_after = df_full[
            (df_full['product_id'] == pair['existing_product_id']) &
            (df_full['date'] >= pair['new_launch'])
        ]['revenue'].sum()

        # Get sales of new product
        new_product_sales = df_full[
            (df_full['product_id'] == pair['new_product_id'])
        ]['revenue'].sum()

        # Calculate sales loss
        if existing_before > 0:
            # Normalize to same time period
            days_before = (pair['new_launch'] - pair['existing_launch']).days
            days_after = (df_full['date'].max() - pair['new_launch']).days

            if days_after > 0 and days_before > 0:
                avg_daily_before = existing_before / days_before
                expected_after = avg_daily_before * days_after
                actual_loss = max(0, expected_after - existing_after)

                # Cannibalization Rate = (Sales Loss / New Product Sales) × 100
                if new_product_sales > 0:
                    cannib_rate = (actual_loss / new_product_sales) * 100
                else:
                    cannib_rate = 0

                cannib_results.append({
                    'brand': pair['brand'],
                    'existing_product': pair['existing_product'],
                    'new_product': pair['new_product'],
                    'sales_before': existing_before,
                    'sales_after': existing_after,
                    'expected_sales_after': expected_after,
                    'sales_loss': actual_loss,
                    'new_product_sales': new_product_sales,
                    'cannibalization_rate': cannib_rate
                })

    cannib_results_df = pd.DataFrame(cannib_results)

    if len(cannib_results_df) > 0:
        # Sort by cannibalization rate
        cannib_results_df = cannib_results_df.sort_values('cannibalization_rate', ascending=False)

        print("\n" + "="*60)
        print("CANNIBALIZATION ANALYSIS RESULTS")
        print("="*60)

        print("\n TOP 10 HIGHEST CANNIBALIZATION CASES:")
        print(cannib_results_df.head(10)[['brand', 'existing_product', 'new_product',
                                           'cannibalization_rate']].to_string(index=False))

        print(f"\n SUMMARY STATISTICS:")
        print(f"Average Cannibalization Rate: {cannib_results_df['cannibalization_rate'].mean():.2f}%")
        print(f"Median Cannibalization Rate: {cannib_results_df['cannibalization_rate'].median():.2f}%")
        print(f"Max Cannibalization Rate: {cannib_results_df['cannibalization_rate'].max():.2f}%")

        # Categorize cannibalization severity
        cannib_results_df['severity'] = pd.cut(
            cannib_results_df['cannibalization_rate'],
            bins=[0, 10, 25, 50, 100],
            labels=['Low', 'Moderate', 'High', 'Severe']
        )

        print("\n CANNIBALIZATION SEVERITY DISTRIBUTION:")
        print(cannib_results_df['severity'].value_counts().to_string())

        # Save results
        cannib_results_df.to_csv('cannibalization_analysis.csv', index=False)
        print("\n Cannibalization results saved to 'cannibalization_analysis.csv'")
    else:
        print("\n No cannibalization data available to analyze")
        cannib_results_df = pd.DataFrame()
```

###BAGIAN 4: MARKETING MIX OPTIMIZATION

Step 8: Marketing Spend Simulation

```python
print("\n" + "="*60)
print("ADVANCED TECHNIQUE 4: MARKETING MIX OPTIMIZATION")
print("="*60)

# Scenario: Simulate impact of increasing marketing spend on revenue

print("\n SIMULATING MARKETING SPEND IMPACT ON REVENUE...")

# Load trained revenue forecasting model
try:
    revenue_model = joblib.load('xgb_revenue_forecasting_model.pkl')
    print(" Revenue forecasting model loaded")
except:
    print(" Using newly trained model")
    revenue_model = xgb_reg

# Create baseline scenario
baseline_data = X_test_scaled.copy()

# Identify marketing-related features
marketing_features = [col for col in X_train.columns if 'marketing' in col.lower() or 'spend' in col.lower()]
print(f"\nMarketing-related features identified: {len(marketing_features)}")
for feat in marketing_features[:10]:
    print(f"  - {feat}")

# Scenario Analysis: Increase marketing spend by 10%, 20%, 30%, 40%, 50%
scenarios = {}
increase_percentages = [0, 10, 20, 30, 40, 50]

for pct in increase_percentages:
    scenario_data = baseline_data.copy()

    # Increase marketing spend features
    for feat in marketing_features:
        if feat in scenario_data.columns:
            feat_idx = list(scenario_data.columns).index(feat)
            scenario_data[:, feat_idx] = scenario_data[:, feat_idx] * (1 + pct/100)

    # Predict revenue
    predicted_revenue = revenue_model.predict(scenario_data)
    scenarios[pct] = predicted_revenue.mean()

print("\n" + "="*60)
print("MARKETING SPEND OPTIMIZATION RESULTS")
print("="*60)

scenario_df = pd.DataFrame({
    'Marketing Spend Increase (%)': list(scenarios.keys()),
    'Predicted Avg Revenue (IDR)': list(scenarios.values())
})

scenario_df['Revenue Increase (%)'] = ((scenario_df['Predicted Avg Revenue (IDR)'] / scenario_df['Predicted Avg Revenue (IDR)'].iloc[0]) - 1) * 100
scenario_df['ROI'] = scenario_df['Revenue Increase (%)'] / scenario_df['Marketing Spend Increase (%)']
scenario_df['ROI'] = scenario_df['ROI'].fillna(0)

print("\n")
print(scenario_df.to_string(index=False))

# Visualize scenarios
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Revenue vs Marketing Spend
axes[0].plot(scenario_df['Marketing Spend Increase (%)'],
             scenario_df['Predicted Avg Revenue (IDR)'],
             marker='o', linewidth=2, markersize=8, color='darkgreen')
axes[0].set_xlabel('Marketing Spend Increase (%)', fontsize=12)
axes[0].set_ylabel('Predicted Avg Revenue (IDR)', fontsize=12)
axes[0].set_title('Marketing Spend vs Predicted Revenue', fontweight='bold', fontsize=14)
axes[0].grid(alpha=0.3)

# ROI by scenario
axes[1].bar(scenario_df['Marketing Spend Increase (%)'][1:],
            scenario_df['ROI'][1:],
            color='coral', alpha=0.7)
axes[1].set_xlabel('Marketing Spend Increase (%)', fontsize=12)
axes[1].set_ylabel('ROI (Revenue Increase / Spend Increase)', fontsize=12)
axes[1].set_title('Marketing ROI by Scenario', fontweight='bold', fontsize=14)
axes[1].axhline(y=1, color='red', linestyle='--', linewidth=2, label='Break-even (ROI=1)')
axes[1].legend()
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

# Find optimal scenario
optimal_scenario = scenario_df.loc[scenario_df['ROI'].idxmax()]

print("\n BUSINESS RECOMMENDATIONS:")
print(f"\n OPTIMAL MARKETING SPEND INCREASE: {optimal_scenario['Marketing Spend Increase (%)']:.0f}%")
print(f"   - Expected Revenue Increase: {optimal_scenario['Revenue Increase (%)']:.2f}%")
print(f"   - ROI: {optimal_scenario['ROI']:.2f}x")
print(f"   - Predicted Avg Revenue: Rp {optimal_scenario['Predicted Avg Revenue (IDR)']:,.0f}")

print("\n INSIGHTS:")
if optimal_scenario['Marketing Spend Increase (%)'] == 0:
    print("   - Current marketing spend is already optimal")
    print("   - Further increases may not yield positive ROI")
else:
    print(f"   - Increasing marketing spend by {optimal_scenario['Marketing Spend Increase (%)']:.0f}% maximizes ROI")
    print(f"   - Beyond this point, diminishing returns set in")
    print(f"   - Expected incremental revenue: Rp {(scenario_df['Predicted Avg Revenue (IDR)'].iloc[-1] - scenario_df['Predicted Avg Revenue (IDR)'].iloc[0]):,.0f}")

# Channel-specific optimization (if channel data available)
if 'channel_name' in df_full.columns:
    print("\n CHANNEL-SPECIFIC RECOMMENDATIONS:")

    # Analyze which channels have best marketing ROI
    if 'marketing_roi' in df_full.columns:
        channel_roi = df_full.groupby('channel_name')['marketing_roi'].mean().sort_values(ascending=False)

        print("\nTop 5 Channels by Marketing ROI:")
        for idx, (channel, roi) in enumerate(channel_roi.head(5).items(), 1):
            print(f"   {idx}. {channel}: {roi:.2f}x ROI")

        print("\n💡 Action: Reallocate budget to high-ROI channels for maximum efficiency")
```
